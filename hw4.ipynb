## Question 8 (10 points)
Compute the average mean daily temperature for each month. Output it as a dictionary

Expected Output
```

 'Mar': 34.00645161290322,
 'Apr': 43.42666666666666,
 'May': 61.4967741935484,

 'Sep': 63.19000000000001,

 'Dec': 30.55806451612903}
 ```

````python
month = a[:, 0].astype(int)
mdt = a[:, 3]
avgMonth = {}

for i in range(1, 13):
    indices = np.where(month == i)[0]
    mdtAvg = np.mean(mdt[indices]) 
    avgMonth[f"{i:02d}"] = mdtAvg

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun","Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
monthlyAvg = {months[int(h) - 1]: i for h, i in avgMonth.items()}
monthlyAvg
````

## Question 9 (15 points)
For each month, count the number of days in the year with mixed precipitation (i.e. both snow and liquid precipation were recorded in the same day). Output as a dictionary

Expected Output
```
{'Jan': 6,
 'Feb': 6,
 'Mar': 4,
 'Apr': 1,
 'May': 0,
 'Jun': 0,
 'Jul': 0,
 'Aug': 0,
 'Sep': 0,
 'Oct': 0,
 'Nov': 4,
 'Dec': 7}
```

````python
month = a[:, 0]
rain = a[:, 6]
snow = a[:, 7]

result = {}

for m in range(1, 13):
    indices = np.where(month == m)[0]
    count = np.sum((rain[indices] > 0) & (snow[indices] > 0))
    month_name = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}[m]
    result[month_name] = count

result
````

## Question 10 (15 points)
Compute and ouput the month that received the highest amount of precipitation (liquid only).Print both the month name and the amount.\
Expected Output
```
('Aug', 3.663)
```

````python
day = a[:,0]
rain = a[:,6]
month = np.zeros(12)
for i in range(12):
    month[i] = np.sum(rain[day == i+1])
rainmax = np.argmax(month)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthlyrain = months[rainmax]
raintotal = month[rainmax]
monthlyrain,raintotal


````

## Question 11 (10 points)
Using Matplotlib, 
1. Plot the maximum, minimum and mean daily temperature for all days on the same graph.
2. The x-axis should reflect the day index
3. The y-axis reflects the temperature values.
4. Add labels to your axes and add a legend and title. (Your graph should have 3 sets of data being plotted)

````python
temp_max = a[:, 1]
temp_min = a[:, 2]
temp_mean = a[:, 3]
days = np.arange(1, 366)
figure, graph = plt.subplots()
graph.plot(days, temp_max, label='Max Temperature', color = "green") #max
graph.plot(days, temp_min, label='Min Temperature', color = "red") #min
graph.plot(days, temp_mean, label='Mean Temperature', color = "yellow") #mean
graph.set_xlabel('Day Index')
graph.set_ylabel('Temperature (Fahrenheit)')
graph.set_title('Daily Temperature in Saginaw, MI (2022)')
graph.legend()
plt.show()
````
